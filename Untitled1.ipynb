```python
# fill null values
data = data.fillna(method='ffill')
```

```python
#check for null
data.isnull().any()
```

```python
# Getting the values and plotting it
#f1 = data['CUST_ID'].values
f2 = data['BALANCE'].values
f3 = data['BALANCE_FREQUENCY'].values
f4 = data['PURCHASES'].values
f5 = data['ONEOFF_PURCHASES'].values
f6 = data['INSTALLMENTS_PURCHASES'].values
f7 = data['PURCHASES_FREQUENCY'].values
f8 = data['ONEOFF_PURCHASES_FREQUENCY'].values
f9 = data['PURCHASES_INSTALLMENTS_FREQUENCY'].values
f10 = data['PURCHASES_TRX'].values
f11 = data['CREDIT_LIMIT'].values
f12 = data['PAYMENTS'].values
f13 = data['MINIMUM_PAYMENTS'].values
f14 = data['PRC_FULL_PAYMENT'].values
f15 = data['TENURE'].values

X = np.array(list(zip(f2, f3, f4, f5, f6,f7,f8,f9,f10,f11,f12,f13,f14,f15)))
X[0:10]
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=4)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_

labels = kmeans.labels_
labels

```

```python
# iteractions needed to set the final centroids
iterations=kmeans.n_iter_
iterations
```

```python
# Comparing with scikit-learn centroids

print(centroids) # From sci-kit learn
```

```python
#print(labels)

colors = ["g.","r.","c.","y."]

for i in range(len(X)):
    print("coordinate:",X[i], "label:", labels[i])
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)


plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)

plt.show()
```

```python
# draw or plot K vs Sum_of_Squared_distance to find the optimal k
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```
